## Process and clean passages

```python
passages = pd.read_pickle('passages.pkl')
passages
```

```python
importlib.reload(twee_utils)

# Rename twee -> passage
if 'twee' in passages:
    passages['passage'] = passages['twee']
    del passages['twee']
print(passages.shape)

passages['original_passage'] = passages.passage

passages['char_count'] = passages.passage.apply(len)

# Cut out the long passages
passages = passages[passages.char_count < 1024]
print('without long passages', passages.shape)

# Clean the twee - for some reason this takes a long time
# Remove images, duplicate newlines, etc
passages['passage'] = passages.passage.progress_apply(twee_utils.clean_twee)

# Split into lines
# TODO for some reason this doubles the number of passages
passages['lines'] = passages.passage.apply(twee_utils.split_lines)
passages['line_count'] = passages.lines.apply(len)
passages['first_line'] = passages.lines.apply(lambda x: (x[0] if x else ''))
if 'lines' in passages:
    del passages['lines']

# Drop Duplicates
passages.drop_duplicates(subset=['passage'])
print(passages.shape)
passages.sample(20)

# Look at macros
passages['macros'] = passages.passage.progress_apply(twee_utils.get_macros)
```

```python
# Look at macros
passages = passages[passages.macros.apply(len) == 0]
print(passages.shape)
```

```python
# Prepare passages for generation
importlib.reload(twee_utils)
pd.set_option('max_colwidth',200)
passages['gen_input'] = passages.passage.apply(twee_utils.twee_to_gen_format)
```
